```python
import cv2
import numpy as np

cap=cv2.VideoCapture("C:\\Users\\siraj\\UB_prototypes\\Yolo projects\\Yolov5\\video_test.mp4")
frame_width = int(cap.get(3)//3)
frame_height = int(cap.get(4)//3)
size = (frame_width, frame_height)

_,first_frame=cap.read()
first_gray=cv2.cvtColor(first_frame,cv2.COLOR_BGR2GRAY)

while True:
    _,frame=cap.read()
    frame=cv2.resize(frame,size)
    first_gray=cv2.resize(first_gray,size)
    gray_frame=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    gray_frame=cv2.resize(gray_frame,size)
    first_gray=cv2.GaussianBlur(first_gray, (5,5), 0)
    gray_frame=cv2.GaussianBlur(gray_frame, (5,5), 0)
    difference=cv2.absdiff(first_gray,gray_frame)
    ret,difference=cv2.threshold(difference, 25, 255, cv2.THRESH_BINARY)
    
    #cv2.imshow("First frame", first_frame)
    #cv2.imshow("Frame", frame)
    cv2.imshow("dife", difference)
    key=cv2.waitKey(25)
    if key==27:
        break
        
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np 

cap=cv2.VideoCapture("C:\\Users\\siraj\\UB_prototypes\\Yolo projects\\Yolov5\\video_test.mp4")
frame_width = int(cap.get(3)//3)
frame_height = int(cap.get(4)//3)
size = (frame_width, frame_height)
subtractor=cv2.createBackgroundSubtractorMOG2(varThreshold = 30)

while True:
    _,frame=cap.read()
    mask=subtractor.apply(frame)
    frame=cv2.resize(frame,size)
    mask=cv2.resize(mask,size)
    contours, hierarchy = cv2.findContours(mask,cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    for c in contours:
        area = cv2.contourArea(c)
        print("area",area)
        if (area>400):
            x,y,w,h = cv2.boundingRect(c)
            cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
    cv2.imshow("Frame",frame)
    cv2.imshow("Mask",mask)
    key=cv2.waitKey(25)
    if key==27:
        break
        
cap.release()
cv2.destroyAllWindows()
```
